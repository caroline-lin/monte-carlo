```python
def ohno(spikiness):
    Z = (X ** 2 + Y ** 2 ) 

    theta = (np.sign(X) - 1.) * -0.5 * np.pi + np.arctan( (Y / X) )
    for i in range(3, 7):
        Z -= (1. / ((X ** 2 + Y ** 2) * (0.5 * i)) ) * np.abs(np.sin(spikiness * np.pi * i * theta))

    return Z
```
